```python
df = pd.read_csv('STATIONS_DATA_.csv')
```

```python
spearman_correlation_matrix = df.drop('Date', axis=1).corr(method='spearman')

plt.figure(figsize=(6, 4))
sns.heatmap(spearman_correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True)
plt.title('Spearman Correlation Matrix of Rain Stations')
plt.show()
```

```python
kendall_correlation_matrix = df.drop('Date', axis=1).corr(method='kendall')

plt.figure(figsize=(6, 4))
sns.heatmap(kendall_correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True)
plt.title('Kendall Correlation Matrix of Rain Stations')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the number of rows and columns for subplots
num_cols = 3  # You want 3 columns
num_rows = (len(df.columns) - 1) // num_cols + 1

# Create the figure and subplots
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(10, 3 * num_rows))

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Loop through columns, skipping "Date"
for i, column in enumerate(df.columns):
    if column != 'Date':
        # Create the displot for the current column on the corresponding subplot
        sns.histplot(df[column], bins=10, kde=True, color='skyblue', ax=axes[i])
        # Set title and adjust layout
        axes[i].set_title(f'Distribution Plot of {column}')
        axes[i].set_xlabel('Rainfall (mm)')
        axes[i].set_ylabel('Density')

# Hide any unused subplots (if the number of columns doesn't divide evenly)
for j in range(i+1, len(axes)):
    axes[j].set_visible(False)

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```
