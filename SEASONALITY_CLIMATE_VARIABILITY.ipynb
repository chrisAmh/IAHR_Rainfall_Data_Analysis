### SEASONAL DECOMPOSITION

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# # Assuming data has already been cleaned from missing values
data = pd.read_csv('Dataset/RAINFALL_FINAL.csv')
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')

# Set 'Date' as the index of the dataframe
data.set_index('Date', inplace=True)

# Decompose the data
result = seasonal_decompose(data, model='additive', period=365)

reconstructed = result.trend + result.seasonal + result.resid


result.plot()
plt.show()

```

```python
reconstructed.plot()
reconstructed.to_csv('new.csv')
```
